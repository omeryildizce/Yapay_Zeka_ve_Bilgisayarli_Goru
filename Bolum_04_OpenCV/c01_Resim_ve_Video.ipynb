```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
resim = cv2.imread("../Images/f1.jpg")
```

```python
resim.shape
```

```python
cv2.imshow("Resim",resim)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
video = cv2.VideoCapture(0)

while True:
    _, frame = video.read()
    if _ == False:
        break

    cv2.imshow("Webcam", frame)
    if cv2.waitKey(1) == 27:
        break

video.release()
cv2.destroyAllWindows()
```

```python
red = resim

red_gray = cv2.cvtColor(red, cv2.COLOR_BGR2GRAY)

cv2.imshow("Orginal", red)
cv2.imshow("Gray", red_gray)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
red = cv2.imread("../Images/f1.jpg")
red = red[175:350, 0:700,:1]


cv2.imshow("Orginal", red)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
red = cv2.imread("../Images/f1.jpg")
red = cv2.resize(red, (700,450))

cv2.imshow("Resize", red)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
