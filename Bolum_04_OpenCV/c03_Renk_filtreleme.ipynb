```python
import cv2
```

```python
live = cv2.VideoCapture(0)

while True:
    _, live_frame = live.read()
    if _ == False:
        break
    live_frame = cv2.cvtColor(live_frame, cv2.COLOR_BGR2HLS)
    cv2.imshow("Live", live_frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

live.release()
cv2.destroyAllWindows()
```

```python
cam = cv2.VideoCapture("../Images/cap.mp4")
low = (21,135,105)
up = (58,255,243)

while True:
    _, frame = cam.read()
    if _ == False:
        break
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, low,up)
    final = cv2.bitwise_and(frame, frame, mask)

    cv2.imshow("frame", frame)
    cv2.imshow("hsv", hsv)
    cv2.imshow("mask", mask)
    cv2.imshow("final", final)
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break

live.release()
cv2.destroyAllWindows()
```

```python
cam = cv2.VideoCapture("../Images/cap.mp4")
low = (21,135,105)
up = (58,255,243)

while True:
    _, frame = cam.read()
    if _ == False:
        break
    frame = cv2.Canny(frame, 100, 50)

    cv2.imshow("frame", frame)

    if cv2.waitKey(10) & 0xFF == ord("q"):
        break

live.release()
cv2.destroyAllWindows()
```
