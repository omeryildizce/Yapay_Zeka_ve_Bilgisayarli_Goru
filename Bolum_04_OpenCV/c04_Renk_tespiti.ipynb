```python
import cv2
```

```python
cam = cv2.VideoCapture("../Images/cap.mp4")
low = (21,135,105)
up = (58,255,243)

while True:
    _, frame = cam.read()
    if _ == False:
        break
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, low,up)
    mask = cv2.bitwise_and(mask, mask, mask)

    mask = cv2.Canny(mask, 100,200)

    contours = cv2.findContours(mask.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_TC89_L1)[-2]
    cv2.drawContours(frame, contours, -1, (255,255,0), 3)

    cv2.imshow("frame", frame)
    cv2.imshow("mask", mask)
    
    if cv2.waitKey(10) & 0xFF == ord("q"):
        break

cam.release()
cv2.destroyAllWindows()
```
