```python
import cv2
import numpy as np
```

```python
red = cv2.imread("../Images/f1.jpg")
red.shape
```

```python
red = cv2.resize(red, (500,500))

yeni_red = cv2.rectangle(red, (100,150), (400,370), (0,0,255),2)

cv2.imshow("resim", yeni_red)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
red = cv2.resize(red, (500,500))

yeni_red = cv2.rectangle(red, (100,150), (400,370), (0,0,255),2)
yeni_red = cv2.putText(yeni_red, "Ferrari", (100,130), cv2.FONT_HERSHEY_SCRIPT_SIMPLEX, 2, (255,255,0))
cv2.imshow("resim", yeni_red)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
