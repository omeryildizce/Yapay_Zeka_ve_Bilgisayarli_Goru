Tensorflow

```python
import tensorflow as tf
```

```python
matrix = tf.constant(4, shape=(1,1), dtype = tf.float32) # sabit ttensör oluşturur
matrix
```

```python
matrix = tf.constant([[1,2,3], [4,5,6]], shape=(2,3))
matrix
```

```python
matrix = tf.eye(3) # birim matris
matrix
```

```python
matrix = tf.ones((4,3))
matrix
```

```python
matrix = tf.zeros((3,2,7))
matrix
```

```python
matrix = tf.random.uniform((2, 2), minval=0, maxval=1)
matrix
```

```python
matrix = tf.random.normal((3,3), mean=0, stddev=1)
tf.cast(matrix, dtype=tf.float64)
```

```python
matrix = tf.range(10)
matrix = tf.range(start=0,limit=10, delta=0.5)
matrix
```
